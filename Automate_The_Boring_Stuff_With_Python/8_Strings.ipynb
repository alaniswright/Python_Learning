## Strings

Escape characters

You can place an r before the beginning quotation mark of a string literal to make it a raw string literal. A raw string makes it easier to enter string values that have backslashes by ignoring all escape character


```python
print('Say hi to Bob\'s mother.')
print("He said, \"Hello!\"")
print("C:\\Users\\Alan")
print(r"C:\Users\Alan")
print("First Line\nSecond Line")
print("Tabbed\tText")
print("Backslash:\\")
print("This string contains a single quote (') and a double quote (\").")
```

```python
while True:
    print('Enter your age:')
    age = input()
    if age.isdecimal():
        break
    print('Please enter a number for your age.')

while True:
    print('Select a new password (letters and numbers only):')
    password = input()
    if password.isalnum():
        break
    print('Passwords can only have letters and numbers.')
```

```python
import pyperclip
text = pyperclip.paste()
original_text = text.split('\n')
for line in original_text:
    line = '* ' + line
text = '\n'.join(original_text)
pyperclip.copy(text)
```
