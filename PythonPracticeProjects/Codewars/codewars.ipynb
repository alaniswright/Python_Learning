````python
import re

def to_camel_case(text):
    words = re.split((r'[-_]+'), text)
    num_of_words = len(words)
    print(num_of_words)
    while num_of_words > 1:
        num_of_words -= 1
        words[num_of_words] = words[num_of_words].capitalize()
    concat_words = "".join(words)
    return concat_words
                      
print(to_camel_case("the-stealth-warrior"))
````

````python
def likes(names):
    if len(names) == 0:
        return "no one likes this"
    if len(names) == 1:
        return f"{names[0]} likes this"
    if len(names) == 2:
        return f"{names[0]} and {names[1]} like this"
    if len(names) == 3:
        return f"{names[0]}, {names[1]} and {names[2]} like this"
    else:
        return f"{names[0]}, {names[1]} and {len(names) - 2} others like this"


print(likes([]))
print(likes(["Peter"]))
print(likes(["Jacob", "Alex"]))
print(likes(["Jacob", "Alex"]))
print(likes(["Max", "John", "Mark"]))
print(likes(["Alex", "Jacob", "Mark", "Max"]))
````

```
[]                                -->  "no one likes this"
["Peter"]                         -->  "Peter likes this"
["Jacob", "Alex"]                 -->  "Jacob and Alex like this"
["Max", "John", "Mark"]           -->  "Max, John and Mark like this"
["Alex", "Jacob", "Mark", "Max"]  -->  "Alex, Jacob and 2 others like this"
```

````python
def letters_to_numbers(s):
    t = 0
    for c in s:
        if c.isalpha():
            if c.islower():
                t += ord(c) - 96
            if c.isupper():
                t += (ord(c) - 64) * 2
        if c.isnumeric():
            t += int(c)
    return t
````

````python
def find_in_array(seq, predicate): 
    for index in range(0, len(seq)):
        if predicate(seq[index], index):
            return index
    return -1
        

true_if_even = lambda value, index: value % 2 == 0
find_in_array([1,3,5,6,7], true_if_even) # --> 3

# list.index(element)
````

## Persistent bugger

https://www.codewars.com/kata/55bf01e5a717a0d57e0000ec/train/python

"""
39 --> 3 (because 3*9 = 27, 2*7 = 14, 1*4 = 4 and 4 has only one digit, there are 3 multiplications)
999 --> 4 (because 9*9*9 = 729, 7*2*9 = 126, 1*2*6 = 12, and finally 1*2 = 2, there are 4 multiplications)
4 --> 0 (because 4 is already a one-digit number, there is no multiplication)
"""

````python
def persistence(n):
    turns = 0
    while n > 9:
        as_string = str(n)
        n = int(as_string[0])
        for i in range(1, len(as_string)):
            n *= int(as_string[i])
        turns += 1
    return turns

print(persistence(39))
print(persistence(999))
print(persistence(4))
````

````python
def sort_array(source_array):
    print("---")
    print(source_array)
    odds = {i:n for i, n in enumerate(source_array) if n % 2 == 1}
    print(odds)
    sorted_keys = sorted(odds.keys())
    print(sorted_keys)
    sorted_values = sorted(odds.values())
    print(sorted_values)
    reordered_odds = zip(sorted_keys, sorted_values)
    print(reordered_odds)
    for key, value in reordered_odds:
        source_array[key] = value
    return source_array
                           
print(sort_array([]))
print(sort_array([]))
print(sort_array([10,-10]))
print(sort_array([1, 111, 11, 11, 2, 1, 5, 0]))
print(sort_array([5, 3, 2, 8, 1, 4]))
print(sort_array([5, 3, 1, 8, 0]))
print(sort_array([]))
print(sort_array([5, 3, 2, 8, 1, 4, 11]))





````
