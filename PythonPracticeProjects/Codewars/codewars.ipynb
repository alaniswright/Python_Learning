```python
import re

def to_camel_case(text):
    words = re.split((r'[-_]+'), text)
    num_of_words = len(words)
    print(num_of_words)
    while num_of_words > 1:
        num_of_words -= 1
        words[num_of_words] = words[num_of_words].capitalize()
    concat_words = "".join(words)
    return concat_words
                      
print(to_camel_case("the-stealth-warrior"))
```
