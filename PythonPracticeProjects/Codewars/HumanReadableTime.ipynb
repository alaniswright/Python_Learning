```python
def make_readable(seconds):
    seconds_remaining = seconds % 60
    minutes = int(seconds / 60)
    minutes_remaining = minutes % 60
    hours = int(minutes / 60)
    
    return f"{hours:02}:{minutes_remaining:02}:{seconds_remaining:02}"




from unittest import TestCase
TestCase().assertEqual(make_readable(0), "00:00:00")
TestCase().assertEqual(make_readable(1), "00:00:01")
TestCase().assertEqual(make_readable(60), "00:01:00")
TestCase().assertEqual(make_readable(3599), "00:59:59")
TestCase().assertEqual(make_readable(3600), "01:00:00")
TestCase().assertEqual(make_readable(86399), "23:59:59")
TestCase().assertEqual(make_readable(86400), "24:00:00")
TestCase().assertEqual(make_readable(359999), "99:59:59")

```
